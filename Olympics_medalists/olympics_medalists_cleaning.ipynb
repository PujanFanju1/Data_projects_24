```python
import pandas as pd

df = pd.read_csv('2024_medalists_all.csv')
```

```python
df.columns.to_list()
```

```python
medalists = df[['medalist_name',
                'medal',
                'country_medal',
                'date_of_birth',
                'place_of_birth',
                'lat',
                'lon',
                'sex_or_gender',
                'event_name',
                'event_part_of_sport']]
```

```python
medalists
```

### The event_name contains 'at the 2024 Summer Olympics –', this is redundant and can be removed.

```python
remove_string = 'at the 2024 Summer Olympics –'

medalists['event_name'] = medalists['event_name'].str.replace('at the 2024 Summer Olympics – ', '', regex=False)
```

```python
medalists
```

### Replace all Null values with 'Unknown' for now as we don't want to drop entire rows yet. 

```python
medalists = medalists.fillna('Unknown')
```

### However, the null values in 'event_part_of_sport' can be filled in if we know the 'event_name'.

```python
missing_part_of_sport = medalists[medalists['event_part_of_sport']=='Unknown'][['event_name','event_part_of_sport']]
```

```python
# medalists['event_name'].unique()
```

```python
# Checking to see all the 'event_part_of_sport' values so we know what the match the event_name with missing values to.
medalists['event_part_of_sport'].unique()
```

```python
missing_part_of_sport['event_name'].unique()
```

```python
# If-elif is only used for single boolean values. To replace certain values, we must use .loc.
medalists.loc[medalists['event_name'].str.contains('Artistic swimming', na=False), 'event_part_of_sport'] = 'synchronized swimming'
medalists.loc[medalists['event_name'].str.contains('cycling', na=False), 'event_part_of_sport'] = 'cycle sport'
```

```python
# medalists
```

```python
# There are no more 'unknowns' in the 'event_part_of_sport' columns.
medalists['event_part_of_sport'].unique()
```

### Next we want to check the datatypes of each column and ensure that they are of the correct type.

```python
# Right now, they are all 'objects'.
medalists.dtypes
```

```python
medalists
```

```python
medalists['medal'] = medalists['medal'].astype('category')
medalists['country_medal'] = medalists['country_medal'].astype('category')
# As there are some athletes for which we do not know the DOB, these rows are filled with 'unknown'.
# These cannot be converted to datetime, so we can use 'coerce'.
medalists['date_of_birth'] = pd.to_datetime(medalists['date_of_birth'], format='%d/%m/%Y', errors='coerce')
medalists['lat'] = pd.to_numeric(medalists['lat'], errors='coerce')
medalists['lon'] = pd.to_numeric(medalists['lon'], errors='coerce')
medalists['sex_or_gender'] = medalists['sex_or_gender'].astype('category')
medalists['event_part_of_sport'] = medalists['event_part_of_sport'].astype('category')
```

```python
medalists['country_medal'].unique()
```

```python
# Checking to see if all the columns are of the correct datatype.
medalists.dtypes
```
