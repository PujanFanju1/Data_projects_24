One-hot encoding is a technique used to convert categorical data into numerical data which machine learning algorithms can process.

One-hot encoding is necessary for the features because we don't want the model to treat categories which don't have
any order to be treated like they do when processing them. However, for the target (output), the random forest
algorithm treats each category as a distinct class so there's no need to apply one-hot encoding to it or convert it to numerical- so I can just keep them as 'coffee A', 'coffee B' and 'coffee C'.


```python
X = final_df.drop(columns='Between Coffee A, Coffee B, and Coffee C which did you prefer?')
Y = final_df['Between Coffee A, Coffee B, and Coffee C which did you prefer?']
```

```python
for col in X.columns:
    X[col] = X[col].astype(int)

X
```

```python
# # Fit a RandomForest model
model = RandomForestClassifier()
model.fit(X, Y)

# # Get feature importances
# Feature Importances in a Random Forest model represent the relative importance of each feature in the model's decision-making process.
importances = model.feature_importances_

# This shows the importance of each feature.
feature_importances_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)
```

```python
feature_importances_df
```

```python
important_features = feature_importances_df[feature_importances_df['Importance']>0.01]['Feature'].to_list()
```

```python
X_final = X[important_features]

# We use a part of the dataset to train the model and another part to test how accurate the model is.
# test_size = 0.4 means that 60% of the data will be used for training the model and 40% will be used for testing the accuracy of the model.
X_train, X_test, y_train, y_test = train_test_split(X_final, Y, test_size=0.4, random_state=0)

# Make the random forest classifier model and fit the training data to it.
model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
# The accuracy is measured as the number of correct predictions over the total number of predictions.
print(f'Accuracy: {accuracy:.2f}')
```

```python
# Apply label encoding to 'age' and one-hot encoding to Gender.
age_order = [
    '<18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '>65 years old'
]

# Create the mapping based on the age order. Enumerate takes the age and the index (0,1,2).
age_mapping = {age: index for index, age in enumerate(age_order)}
age_mapping

all_v3_imputed['What is your age?'] = all_v3_imputed['What is your age?'].map(age_mapping)
all_v3_imputed
```

```python
# Encode the gender column using one-hot encoding as this is non-ordinal data.
gender_encoded = pd.get_dummies(all_v3_imputed['Gender'])


# Concat back onto the original df.
all_v3_imputed = pd.concat([all_v3_imputed,gender_encoded],axis=1)
all_v3_imputed.drop(columns='Gender',inplace=True)
```
