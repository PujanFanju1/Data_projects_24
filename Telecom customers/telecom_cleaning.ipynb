```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Telecom Customers Churn.csv')
```

```python
cols = df.columns.to_list()
```

```python
# Checking to see the unique values of each column.
for col in cols:
    print(col)
    print(df[col].unique())
    print(df[col].isnull().sum())
    print(df[col].dtype)
    print('')
```

```python
# TotalCharges seems so have some blank values, so we will replace them with null and then convert from object to numerical.
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Most of the columns are categorical, so we will make a list of all the non-categorical columns.
non_categorical_cols = ['customerID', 'tenure', 'MonthlyCharges', 'TotalCharges']

# Convert all the columns not in the non-categorical columns list to categorical.
categorical_cols = [col for col in df.columns if col not in non_categorical_cols]
df[categorical_cols] = df[categorical_cols].astype('category')
```

```python
# Re-checking to see the unique values of each column.
for col in cols:
    print(col)
    print(df[col].unique())
    print(df[col].isnull().sum())
    print(df[col].dtype)
    print('')
```

```python
# Saving the cleaned dataset to be used for analysis.
df.to_csv('Telecom_data_cleaned.csv')
```
